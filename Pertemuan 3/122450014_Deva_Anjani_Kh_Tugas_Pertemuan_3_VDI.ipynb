## **Visualisasi Data dengan Library dalam Python**

```python
# import library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import dataset
horror_movies = pd.read_csv('Horror.csv')

# menampilkan dataset
horror_movies
```

## **Bagaimana menentukan rekomendasi yang sesuai?**

Saya tidak akan menggunakan semua atribut dalam menentukan rekomendasi, karena menurut saya tidak semua atribut memiliki pengaruh terhadap rekomendasi yang ingin saya berikan. Oleh karena itu, di sini saya akan memberikan rekomendasi kepada Bunda saya berdasarkan atribut 'rating', 'num_raters', dan 'num_reviews'.

##### **Berdasarkan 'rating'**

```python
# memilih 10 film dengan rating tertinggi
recommend1 = horror_movies.nlargest(10, 'rating')

# mengatur ukuran plot
plt.figure(figsize=(8, 6))

# diagram batang untuk atribut 'name' dan 'rating' dengan palet warna revolver
sns.barplot(data=recommend1, x='rating', y='name', hue='rating', palette='light:#301B3F') # Warna: Revolver
plt.title("10 Film Horor Teratas Berdasarkan Rating", fontsize=16)
plt.xlabel("Rating", fontsize=12)
plt.ylabel("Nama Film", fontsize=12)

# menampilkan plot
plt.show()
```

##### **Berdasarkan 'num_raters'**

```python
# memilih 10 film dengan jumlah rater tertinggi
recommend2 = horror_movies.nlargest(10, 'num_raters')

# mengatur ukuran plot
plt.figure(figsize=(8, 6))

# diagram batang untuk atribut 'name' dan 'num_raters' dengan palet warna oxford blue
sns.barplot(data=recommend2, x='num_raters', y='name', hue='num_raters', palette='light:#3C415C') # Warna: Oxford Blue
plt.title("10 Film Horor Teratas Berdasarkan Jumlah Raters", fontsize=16)
plt.xlabel("Jumlah Raters", fontsize=12)
plt.ylabel("Nama Film", fontsize=12)

# menampilkan plot
plt.show()
```

##### **Berdasarkan 'num_reviews'**

```python
# memilih 10 film dengan jumlah ulasan tertinggi
recommend3 = horror_movies.nlargest(10, 'num_reviews')

# mengatur ukuran plot
plt.figure(figsize=(8, 6))

# diagram batang untuk atribut 'name' dan 'num_reviews' dengan palet warna martini
sns.barplot(data=recommend3, x='num_reviews', y='name', hue='num_reviews', palette='light:#B4A5A5') # Warna: Martini
plt.title("10 Film Horor Teratas Berdasarkan Jumlah Ulasan", fontsize=16)
plt.xlabel("Jumlah Ulasan", fontsize=12)
plt.ylabel("Nama Film", fontsize=12)

# menampilkan plot
plt.show()
```
